```
def print_files():
  print(os.walk('./Data/input'))
  for dirname, _, filenames in os.walk('./Data/input'):
    for filename in filenames:
      print(os.path.join(dirname,filename))

def generate_sample_data():
  number_of_rows = 20
  num_cols = 7
  cols = list("ABCDEFG")
  df = pd.DataFrame(np.random.randint(1,20,size=(number_of_rows, num_cols)), columns=cols)
  df.index = pd.util.testing.makeIntIndex(number_of_rows)
  return df

def generate_sample_data_datetime():
  number_of_rows = 366*24
  num_cols = 2
  cols = ["sales","customers"]
  df = pd.DataFrame(np.random.randint(1,20,size=(number_of_rows, num_cols)), columns=cols)
  df.index = pd.util.testing.makeDateIndex(number_of_rows, freq="H")
  return df
```

## Trick 1: Add a prefix or suffix to all columns

```
df = generate_sample_data()[:5]
print("Original df")
display(df)

print("Add prefix")
display(df.add_prefix("1_"))

print("Add suffix")
display(df.add_suffix("_Z"))
```

## Trick 2: Reverse order of a df

```
print("Reverse columns order")
display(df.loc[:, ::-1])

print("Reverse index order")
display(df.loc[::-1])

print("Reverse row order and reset index")
display(df.loc[::-1].reset_index(drop=True))
```

## Trick 3: Filter a df by multiple conditions (isin and iverse using ~)

```
df = generate_sample_data()[:5]
df["A"] = np.arange(5)+1
display(df)

display(df[(df["A"]==1) | (df["A"]==3)])

display(df[df["A"].isin([1,3])])

display(df[~df["A"].isin([1,3])])
```

## Trick 4: Select columns by dtype

```
df = generate_sample_data_datetime()[:10].reset_index()
df["string_col"] = list("ABCDEABCDE")
df["sales"] = df["sales"].astype("float")
display(df)

display(df.info())

display(df.select_dtypes(include="number"))

display(df.select_dtypes(include="object"))

display(df.select_dtypes(include=["datetime","timedelta"]))

display(df.select_dtypes(include=["float"]))
```
